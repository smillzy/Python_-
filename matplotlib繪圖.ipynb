# 3-1 子圖plt.axes

plt.axes(left, bottom, width, height)
- left：子圖左邊位置，範圍0~1
- bottom：子圖下面位置，範圍0~1
- width：圖的寬度，範圍0~1
- height：圖的高度，範圍0~1
- plt.axes(left, bottom, width, height) (左、下、寬、高)，範圍從左下角的0到右上角的1。
- plt.axes([0.65, 0.65, 0.2, 0.2]) 從軸的寬度65%和圖的高度65%開始繪圖，往右上角繪製寬度20%與高度20%的子圖
語法：https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axes.html

# 網格圖 plt.subplot
- 建立的網格子圖行數、列數和索引值，索引值從 1 開始，從左上角到 右下角依次增加編號
- 一次僅產生一個子圖

plt.subplot(nrows, ncols, index)
- nrows：子圖產生的row數
- ncols：子圖產生的col數
- index：子圖編號
- 寫法：plt.subplot(1, 1, 1)或plt.subplot(111)
- 語法：https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html

```python
%pylab inline
x1 = np.arange(0, 360, 5)
y1 = np.sin(x1 * np.pi/180.0)
x2 = x1
y2 = np.cos(x2 * np.pi / 180.0)

plt.subplot(231)
plt.plot(x1, y1, lw = 3)
plt.xlim(-30, 390)
plt.ylim(-1.5, 1.5)

plt.subplot(232)
plt.plot(x2, y2, 'ro')
plt.xlim(-10, 360)
plt.ylim(-1.2, 1.2)

plt.subplot(233)
plt.plot(x2, y2, 'ko')
plt.xlim(0, 50)
plt.ylim(-1.2, 1.2)

plt.subplot(234)
plt.plot(x2, y2, 'b+')
plt.xlim(0, 180)
plt.ylim(-1.2, 1.2)

plt.subplot(236)
plt.plot(x2, y2, 'g*')
plt.xlim(180, 360)
plt.ylim(-1.2, 1.2)
```

# fig.add_subplot：另一個與plt.subplot 類似功能子圖設定
- plt.subplot 會刪除已存在的子圖
- fig.add_subplot 會保留子圖及軸

```python
import matplotlib.pyplot as plt
fig = plt.figure()
fig.subplots_adjust(hspace = 0.4, wspace = 0.4) #調整子圖之間的間隔
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)), fontsize = 18, ha = 'center')
plt.show()
```

# 複雜排列方式plt.GridSpec
- 可產生不規則的多行多列子圖。

plt.GridSpec(nrows, ncols, wspace, hspace)
- nrows：指定Row數
- ncols：指定Column數
- wspace：兩個子圖的寬度距離
- hspace：兩個子圖的高度距離
- 語法：https://matplotlib.org/stable/api/_as_gen/matplotlib.gridspec.GridSpec.html

```python
import matplotlib.pyplot as plt
grid = plt.GridSpec(4, 5, wspace = 0.4, hspace = 0.3)
plt.figure(figsize = (12, 10))
plt.subplot(grid[0, 0])
plt.text(0.5, 0.5, str((0, 0)), fontsize=18, ha='center')
plt.subplot(grid[0, 1])
plt.text(0.5, 0.5, str((0, 1)), fontsize=18, ha='center')
plt.subplot(grid[0, 2])
plt.text(0.5, 0.5, str((0, 2)), fontsize=18, ha='center')
plt.subplot(grid[0, 3])
plt.text(0.5, 0.5, str((0, 3)), fontsize=18, ha='center')
plt.subplot(grid[1, :1])
plt.text(0.5, 0.5, str('1, :1'), fontsize=18, ha='center')
plt.subplot(grid[1, 1:4])
plt.text(0.5, 0.5, str('1,1:4'), fontsize=18, ha='center')
plt.subplot(grid[2, :2])
plt.text(0.5, 0.5, str('2, :2'), fontsize=18, ha='center')
plt.subplot(grid[2, 2:4])
plt.text(0.5, 0.5, str('2, 2:4'), fontsize=18, ha='center')
plt.subplot(grid[3, :3])
plt.text(0.5, 0.5, str('3, :3'), fontsize=18, ha='center')
plt.subplot(grid[3, 3:4])
plt.text(0.5, 0.5, str('3, 3:4'), fontsize=18, ha='center')
plt.subplot(grid[:3, 4:])
plt.text(0.5, 0.5, str(':3, 4:'), fontsize=18, ha='center')
plt.subplot(grid[3:, 4])
plt.text(0.5, 0.5, str('3:, 4'), fontsize=18, ha='center')
plt.show()
```

# subplot2grid

plt.subplot2grid(shape, loc, rowspan=1, colspan=1)
- shape：指定形狀(row, column)
- loc：指定位置(row, column)
- rowspan：row合併
- colspan：column合併
- 語法：https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot2grid.html

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = plt.subplot2grid((3,3), (0,0), colspan=2, rowspan=2) # topleft
ax1.text(0.5, 0.5, str('0,0'), fontsize=18, ha='center')
ax3 = plt.subplot2grid((3,3), (0,2), rowspan=3)            # right
ax3.text(0.5, 0.5, str('0,2'), fontsize=18, ha='center')
ax4 = plt.subplot2grid((3,3), (2,0))                       # bottom left
ax4.text(0.5, 0.5, str('2,0'), fontsize=18, ha='center')
ax5 = plt.subplot2grid((3,3), (2,1))                       # bottom right
ax5.text(0.5, 0.5, str('2,1'), fontsize=18, ha='center')
fig.tight_layout()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Setup the subplot2grid Layout
fig = plt.figure(figsize=(10, 5))
ax1 = plt.subplot2grid((2,4), (0,0))
ax2 = plt.subplot2grid((2,4), (0,1))
ax3 = plt.subplot2grid((2,4), (0,2))
ax4 = plt.subplot2grid((2,4), (0,3))
ax5 = plt.subplot2grid((2,4), (1,0), colspan=2)
ax6 = plt.subplot2grid((2,4), (1,2))
ax7 = plt.subplot2grid((2,4), (1,3))

# Input Arrays
n = np.array([0,1,2,3,4,5])
x = np.linspace(0,5,10)
xx = np.linspace(-0.75, 1., 100)

# Scatterplot
ax1.scatter(xx, xx + np.random.randn(len(xx)))
ax1.set_title("Scatter Plot")

# Step Chart
ax2.step(n, n**2, lw=2)
ax2.set_title("Step Plot")

# Bar Chart
ax3.bar(n, n**2, align="center", width=0.5, alpha=0.5)
ax3.set_title("Bar Chart")

# Fill Between
ax4.fill_between(x, x**2, x**3, color="steelblue", alpha=0.5);
ax4.set_title("Fill Between");

# Time Series
dates = pd.date_range('2018-01-01', periods = len(xx))
ax5.plot(dates, xx + np.random.randn(len(xx)))
ax5.set_xticks(dates[::30])
ax5.set_xticklabels(dates.strftime('%Y-%m-%d')[::30])
ax5.set_title("Time Series")

# Box Plot
ax6.boxplot(np.random.randn(len(xx)))
ax6.set_title("Box Plot")

# Histogram
ax7.hist(xx + np.random.randn(len(xx)))
ax7.set_title("Histogram")

fig.tight_layout()
```

# 區域填滿

plt.fill(x, y, color, alpha)
- x：指定填滿的x區域
- y：指定填滿的y區域
- color：指定填滿顏色。
- alpha：透明度0~1
- 語法：https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill.html

```python
%pylab inline

x = np.linspace(0, 10)
y1 = x * np.sin(x)
y2 = x * np.cos(x)

plt.figure(figsize=(5,3))
plt.fill(x, y1, 'red', alpha=0.4)
plt.fill(x, y2, 'blue', alpha=0.4)
plt.xlabel('x axis yo!')
plt.ylabel("I don't even know")
```

plt.fill_between(x, y1, y2, where= y1 >= y2, facecolor = "red", interpolate= True, alpha = 0.5)
- x：表示覆蓋的區域
- y1：表示覆蓋的下限
- y2：表示覆蓋的上限
- where：過律條件，指定符合的情況進行繪圖
- facecolor：覆蓋區域的顏色
- interpolate：自動填充空白，當x取得離散點差距較大時，顯示的時候兩個函數之間的區域可能有空白存在，interpolate 就是用來填充這部分區域
- alpha：覆蓋區域的透明度[0,1],其值越大，表示越不透明

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5 * np.pi, 1000)

y1 = np.sin(x)
y2 = np.sin(2 * x)

plt.figure(figsize=(15,5))
plt.plot(x, y1, c = "g")
plt.plot(x, y2, c = 'r')

plt.fill_between(x, y1, y2, facecolor = "yellow")
```
