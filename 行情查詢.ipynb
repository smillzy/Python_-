# 計算每日報酬率計算方法分3種
- 方法1
$$ 報酬率 = \frac{期末-期初}{期初}$$

- 方法2
$$ 報酬率 = (\frac{期末}{期初} - 1 )* 100$$

- 方法3： pandas提供的 .pct_change()

```python
#方法1：
((df['Adj Close'] - df['Adj Close'].shift(1)) / df['Adj Close'].shift(1) ).head(10)
```

```python
#方法2
(df['Adj Close'] / df['Adj Close'].shift(1) - 1).head(10)
```

```python
#方法3
df['Adj Close'].pct_change().head(10)
```

```python
df['Daily Return'] = df['Adj Close'].pct_change()
df
```

# 每日報酬繪圖

```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
df['Daily Return'].plot()
```

# 移動視窗 .rolling() ，計算移動平均線

```python
#計算移動5日平均線
df['MA5'] = df['Close'].rolling(window = 5).mean()
#計算移動10日平均線
df['MA10'] = df['Close'].rolling(window = 10).mean()
#計算移動20日平均線
df['MA20'] = df['Close'].rolling(20).mean()
df
```

```python
#計算移動5日平均線
df['max5'] = df['Close'].rolling(window = 5).max()
#計算移動10日平均線
df['min10'] = df['Close'].rolling(window = 10).min()
df
```

```python
%pylab inline
df[['Close', 'MA5', 'MA10', 'MA20']].plot(figsize = (15, 8))
```

```python
from FinMind.data import DataLoader

stock_no = '0050'
dl = DataLoader()
stock_data = dl.taiwan_stock_daily(stock_id=stock_no, start_date='2000-01-01')
stock_data.head()
```

```python
import pandas as pd

stock_data.set_index("date" , inplace=True)
stock_data = stock_data.set_index(pd.DatetimeIndex(pd.to_datetime(stock_data.index)))
stock_data
```

```python
stats = stock_data[["close"]].calc_stats()
stats.display()
```
