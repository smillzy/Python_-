```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(10)
df = pd.DataFrame({
    'col1':np.random.randint(10, size = 5),
    'col2':np.random.randint(10, size = 5)
})
df
```

```python
df.plot(figsize = (5, 3), kind = 'bar', rot = 0) #rot全名rotation
```

```python
df.plot(figsize = (5, 3), 
        xlim = (0, 8),
        ylim = (0,15),
        rot = 0) #rot全名rotation
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Microsoft Yahei"
df.plot(figsize = (5, 3), 
        rot = 0,
        marker = "v",
        markersize = 16,
        markerfacecolor = "yellow", 
        markeredgecolor = "#62def7",
       linewidth = 3,
       color = ["r", "g"],
       title = "業務課業績",
       xlabel = "月份",
       ylabel = "業績"
       )
plt.savefig("業務課業績.png",format = "png", bbox_inches = "tight", dpi = 200)
```

# 有加S - plt.subplots

```python
fig, ax = plt.subplots(2, 3, figsize = (8 , 5))
ax[0, 0].plot(df['col1'], color = 'r')
ax[0, 0].set_ylabel('ax[0, 0]Y軸標題')
ax[0, 0].set_title('ax[0, 0]圖形大標題')
ax[0, 0].text(2, 5, '折線圖', ha = 'center', color = 'g')

ax[0, 1].bar(df.index, df['col1'], color = 'g')
ax[0, 1].set_ylabel('ax[0, 1]Y軸標題')
ax[0, 1].set_title('ax[0, 1]圖形大標題')
ax[0, 1].text(2, 5, '折線圖', ha = 'center', color = 'r')

ax[0, 2].bar(df.index, df['col1'], color = 'y')
ax[0, 2].bar(df.index, df['col2'], color = 'r', bottom = df['col1'])
ax[0, 2].set_ylabel('ax[0, 2]Y軸標題')
ax[0, 2].set_title('ax[0, 2]圖形大標題')
ax[0, 2].text(2, 5, '堆疊長條圖', ha = 'center', color = 'k')

width = 0.3
ax[1][0].bar(df.index, df['col1'], color = 'y')
ax[1][0].bar(df.index + df['col2'], width, color = 'r')
ax[1][0].set_ylabel('ax[1][0]Y軸標題')
ax[1][0].set_title('ax[1][0]圖形大標題')
ax[1][0].text(2, 5, '堆疊長條圖', ha = 'center', color = 'k')

ax[1][1].scatter(df.index, df['col1'], s = 100)
ax[1][1].set_ylabel('ax[1][1]Y軸標題')
ax[1][1].set_title('ax[1][1]圖形大標題')
ax[1][1].text(2, 5, '散點圖', ha = 'center', color = 'k')

df1 = pd.DataFrame({0:{'2330':92, '3008':63, '2454':81, '3661':75}})
ax[1][2].pie(df1[0], labels = df1.index, autopct = '%.1f%%')
ax[1][2].set_ylabel('ax[1][2]Y軸標題', labelpad = 20)
ax[1][2].set_title('ax[1][2]圖形大標題')

plt.subplots_adjust(hspace = 0.4, wspace = 0.4)
plt.suptitle('2330財報')

plt.savefig("2330財報.png",format = "png", bbox_inches = "tight", dpi = 200)
```

# 沒加S - plt.subplot(row, column, number)

```python
plt.figure(figsize = (10 , 5))
plt.subplot(2, 3, 1)
plt.plot(df['col1'], color = 'r')
plt.ylabel('231 Y軸標題')
plt.title('231圖形大標題')
plt.text(2, 5, '折線圖', ha = 'center', color = 'g')

plt.subplot(2, 3, 2)
plt.bar(df.index, df['col1'], color = 'g')
plt.ylabel('232 Y軸標題')
plt.title('232圖形大標題')
plt.text(2, 5, '折線圖', ha = 'center', color = 'r')

plt.subplot(2, 3, 3)
plt.bar(df.index, df['col1'], color = 'y')
plt.bar(df.index, df['col2'], color = 'r', bottom = df['col1'])
plt.ylabel('233 Y軸標題')
plt.title('233圖形大標題')
plt.text(2, 5, '堆疊長條圖', ha = 'center', color = 'k')

plt.subplot(234)
width = 0.3
plt.bar(df.index, df['col1'], color = 'y')
plt.bar(df.index + df['col2'], width, color = 'r')
plt.ylabel('234 Y軸標題')
plt.title('234圖形大標題')
plt.text(2, 5, '堆疊長條圖', ha = 'center', color = 'k')

plt.subplot(235)
plt.scatter(df.index, df['col1'], s = 100)
plt.ylabel('235 Y軸標題')
plt.title('235圖形大標題')
plt.text(2, 5, '散點圖', ha = 'center', color = 'k')

plt.subplot(236)
df1 = pd.DataFrame({0:{'2330':92, '3008':63, '2454':81, '3661':75}})
plt.pie(df1[0], labels = df1.index, autopct = '%.1f%%')
plt.ylabel('236 Y軸標題', labelpad = 20)
plt.title('236圖形大標題')

plt.subplots_adjust(hspace = 0.4, wspace = 0.4)
plt.suptitle('2330財報')

plt.savefig("2330財報.png",format = "png", bbox_inches = "tight", dpi = 200)
```
